```python
import pandas as pd
```

```python
#import file
job_data = pd.read_csv('Job_Data.csv')
job_data.head(20)
```

```python
job_data = job_data.dropna(how="any")
```

```python
def split_record(location):
    ret_val = location
    if ("NY" in location):
        ret_val = location.split(", NY")[0]
    else:
        if "CA" in location:
            ret_val = location.split(", CA")[0]
    return ret_val
```

```python
job_data["City"] = job_data["Location"].map(split_record)
```
